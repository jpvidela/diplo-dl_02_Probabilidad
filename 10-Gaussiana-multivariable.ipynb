# Gaussiana Multivariable

Dado un vector aleatoria con distribución gaussiana. Esta queda determinada por el vector de medias y la matriz de covarianza

$\vec{X} = [X_1, X_2, X_3, ..., X_N]$

$\Sigma_{\vec{X}} = 
\begin{bmatrix}
    \sigma_{x_1}^2 & cov(x_1,x_2) & cov(x_1,x_3) & \dots  & cov(x_1,x_N) \\
    cov(x_2,x_1) & \sigma_{x_2}^2 & cov(x_2,x_3) & \dots  & cov(x_2,x_N) \\
    \vdots & \vdots & \vdots & \ddots & \vdots \\
    cov(x_N,x_1) & cov(x_N,x_2) & cov(x_N,x_3) & \dots  & \sigma_{x_N}^2
\end{bmatrix}$

```python
import numpy as np
```

```python
np.array([0, 0])
```

```python
np.array([[1,0],[0,1]])
```

La desviación estándar de ambas variables es $ \sqrt{1} = 1$. Además se observa que no hay correlación entre las variables $Cov(x_1,x_2) = Cov(x_2,x_1) = 0$

Modificando los valores de media y varianza en el ejemplo debajo se observa que:
- Una variación de la media genera un desplazamiento del centro de la campana.
- Una variación de los elementos de la diagonal de la matriz de covarianzas, genera un ensanchamiento/estrechamiento de la campana según el valor de la diagonal que se altere
- Los demás elementos de la matriz de covarianza representan la fuerza de la correlación entre ambas variables, por lo que mientras mayor resulten, 
- La covarianza no puede ser mayor que la varianza de la variable

```python
medias = np.array([0,0])
varianza = np.array([[5,0],[0,1]])

plot_mv_gaussian(medias, varianza,  autoscale=([-2, 2], [-2, 2], [0, 0.15]))
```
